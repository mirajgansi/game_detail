```python

df_uniqune.loc[:,'highest_price'] = df['highest_price'].fillna('free')
```

```python
df_uniqune = df_uniqune.dropna(subset=['highest_price'])
```

```python
row_cols=['metacritic_score', 'metacritic_rating_count','metacritic_user_score','metacritic_user_rating_count','playstation_rating_count'] 
for cols in row_cols:
    df_uniqune[cols] = df_uniqune[cols].fillna(method='ffill')
```

```python
df_uniqune[df_uniqune["publisher"].isnull()]
```

```python
df_uniqune['publisher']= df['publisher'].fillna('unknown')
```

```python

free= df_uniqune[df_uniqune['highest_price']==0]
print(len(free))
```

```python
df_uniqune['highest_price'] = (
    df_uniqune['highest_price']
    .replace('free', '0', regex=True)                     # replace "free" with "0"
    .replace('[\€\$\£\,]', '', regex=True)                # remove symbols
)

df_uniqune['highest_price'] = pd.to_numeric(df_uniqune['highest_price'], errors='coerce').fillna(0).round().astype(int)
```

```python
summary = df_uniqune[['game_name', 'highest_price', 'metacritic_user_rating_count']]
summary
```

```python
summary.sort_values(by='highest_price', ascending=True)
```

```python
summary['highest_price'] = pd.to_numeric(summary['highest_price'], errors='coerce')

# Compute correlation between two columns
corr = summary['highest_price'].corr(summary['metacritic_user_rating_count'])
print('Correlation:', corr)

# Compute correlation matrix
corr_matrix = summary.select_dtypes(include=['number']).corr()

# Create heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Game Price and User Counts Correlation Heatmap")
plt.show()
```

```python
bins=  [0, 100, 200, 400, 600, 800, 1000, 2000, 5000, 10000]
summary['user_bins']= pd.cut(summary['metacritic_user_rating_count'],bins)

plt.figure(figsize=(10,6))
sns.boxplot(x='price_bin', y='metacritic_user_rating_count', data=summary)
plt.title('Boxplot of Ratings by Price')
plt.xlabel('Highest Price')
plt.ylabel('User Rating Count')
plt.show()
```
